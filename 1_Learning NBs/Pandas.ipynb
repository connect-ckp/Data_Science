<h2>set_index(), reset_index(), sort_index()</h2>

`set_index():` Set the DataFrame index (row labels) using one or more existing columns or arrays (of the correct length). The index can replace the existing index or expand on it.<br>

`reset_index():`: Resets index to original<br>

`sort_index():` Sorts object by labels (along an axis)

```python
df.head()
```

```python
#Since Id is unique for every row, set Id as index of the DF. To make the change inplace, use inplace = True
df.set_index('Id', inplace = True)
df
```

```python
df.index
```

```python
df.reset_index(inplace = True)
df
```

<h2>iloc & loc</h2>
`iloc` - Access rows by integer location (Column names become the index)<br>
`loc`  - Access rows by labels or index

**`iloc:`**

```python
df.head()
```

```python
df.iloc[3] #3rd Row
```

```python
df.iloc[5][['RoofMatl', 'Exterior1st', 'Exterior2nd']]
```

```python
df.iloc[0,1]
```

```python
df.iloc[[0,1,2]] #Returns DF of Rows-0,1,2
```

```python
df.iloc[[0,1,2],[2,3,4]] #Returns DF of Rows-0,1,2 and cols-2,3,4
```

```python
df.iloc[0:3, 2:5] #Same through slicing
```

**`loc:`**

```python
df.head()
```

```python
df.loc[3] # Row having index = 3
```

```python
df.loc[[0,1,2],['MSZoning','LotFrontage','LotArea']]
```

```python
df.loc[0:2,'MSZoning':'LotArea'] #Note that in loc, slicing is inclusive at both sides
```

```python
df.set_index('Id',inplace=True)
```

```python
df.head()
```

```python
df.loc[3] # Row having index = 3, compare this with df.iloc[3] below.
```

```python
df.iloc[3]
```

<h2>Applying Filters</h2>

```python
df['Street'] == 'Grvl'
```

```python
filter_cond = (df['Street'] == 'Grvl')
filter_cond
```

```python
df[filter_cond]
```

```python
# Doing the same directly
df[df['Street'] == 'Grvl']
```

```python
#Applying filters through df.loc
df.loc[filter_cond]
```

```python
df.loc[filter_cond, ['MSZoning','LotFrontage','LotArea']]
```
